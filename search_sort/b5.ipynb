```python
def binary_search(l, v, left=None, right=None):
    left = left or 0
    right = right or len(l) - 1
    while left <= right:
        mid = (left + right) // 2
        elem = l[mid]
        if elem == v:
            return mid
        if v < elem:
            right = mid - 1
        else:
            left = mid + 1

def linear_search(l, v, start=None, stop=None):
    for index in range(start or 0, stop or len(l)):
        element = l[index]
        if element == v:
            return index

def create_series(list, algo, count=10):
    increment = len(list) // count
    for index in range(1, len(list), increment):
        yield index - 1, stop_watch(list, list[index-1], algo, 0, index)

def stop_watch(list, value, algo, start_index=None, stop=None):
    import time
    start = time.time()
    algo(list, value, start_index, stop)
    return time.time() - start
```

```python
from null79 import names

binary = list(create_series(names, binary_search))
print(binary)

linear = list(create_series(names, linear_search))
print(linear)

```
